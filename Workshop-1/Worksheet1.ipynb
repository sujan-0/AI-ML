**PCA**

```python
import numpy as np
import matplotlib . pyplot as plt
from PIL import Image

image = Image.open("/content/drive/MyDrive/AI&ML-Level6/unnamed.png").convert("L")
image_array = np . array ( image )
height , width = image_array.shape
data = image_array.reshape( height , width )
plt.imshow(data, cmap = "gray")
```

```python
# Compute the mean of each column ( feature )
mean = np . mean ( data , axis =0)
# Subtract mean to center the data
centered_data = data - mean
```

```python
# Compute covariance matrix
cov_matrix = np . cov ( centered_data , rowvar = False )
```

```python
import numpy as np

# Step 2: Compute eigenvalues and eigenvectors
# We use np.linalg.eigh because it is optimized for symmetric (or Hermitian) matrices like a covariance matrix
eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)

# Step 3: Sort eigenvalues and corresponding eigenvectors in descending order
# We use np.argsort to get the indices of the sorted eigenvalues
# The [:: -1] reverses the order, so that we get the indices in descending order
sorted_indices = np.argsort(eigenvalues)[:: -1]

# Step 4: Reorder eigenvalues to match the sorted order
eigenvalues = eigenvalues[sorted_indices]

# Step 5: Reorder eigenvectors to match the sorted order of eigenvalues
# We apply the sorted indices to the eigenvectors, since eigenvectors are column vectors
eigenvectors = eigenvectors[:, sorted_indices]
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 2: Compute the explained variance ratio
explained_variance_ratio = eigenvalues / np.sum(eigenvalues)

# Step 3: Plot the cumulative sum of explained variance ratio
plt.plot(np.cumsum(explained_variance_ratio))

# Step 4: Add title to the plot
plt.title("Cumulative Explained Variance")

# Step 5: Label the x-axis
plt.xlabel("Number of Components")

# Step 6: Label the y-axis
plt.ylabel("Cumulative Explained Variance")

# Step 7: Add grid to the plot
plt.grid(True)

# Step 8: Show the plot
plt.show()
```

```python
k = 50 # Choose k principal components
components = eigenvectors [: , : k ]
```

```python
compressed_data = np . dot ( centered_data , components )
```

```python
#Reconstruct the image from compressed data
decompressed_data = np . dot ( compressed_data , components . T ) + mean
```

```python
import matplotlib.pyplot as plt

# Step 1: Create a figure with specific dimensions
plt.figure(figsize=(12, 6))

# Step 2: Display the original image
plt.subplot(1, 3, 1)  # 1 row, 3 columns, 1st subplot
plt.imshow(image_array, cmap="gray")  # Display original image
plt.title("Original Image")
plt.axis("off")  # Remove axes

# Step 3: Display the compressed representation
plt.subplot(1, 3, 2)  # 1 row, 3 columns, 2nd subplot
plt.imshow(compressed_data, cmap="gray", aspect="auto")  # Display compressed image
plt.title(f"Compressed Image (k={k})")  # k is the number of components used in compression
plt.axis("off")  # Remove axes

# Step 4: Display the decompressed image
plt.subplot(1, 3, 3)  # 1 row, 3 columns, 3rd subplot
plt.imshow(decompressed_data, cmap="gray")  # Display decompressed image
plt.title("Decompressed Image")
plt.axis("off")  # Remove axes

# Step 5: Tighten layout to avoid overlap
plt.tight_layout()

# Step 6: Show the images
plt.show()
```
